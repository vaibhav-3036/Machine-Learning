```python
import numpy as np
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
from sklearn.linear_model import Ridge, Lasso , ElasticNet
```

```python
x, y = load_diabetes(return_X_y=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
ridge = Ridge(alpha=0.1)
ridge.fit(x_train, y_train)
y_pred_ridge = ridge.predict(x_test)
print("Ridge R2 score:", r2_score(y_test, y_pred_ridge))
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(x_train, y_train)
y_pred_lasso = lasso.predict(x_test)
print("Lasso R2 score:", r2_score(y_test, y_pred_lasso))
```

```python
elastic_net = ElasticNet(alpha=0.001, l1_ratio=0.5)
elastic_net.fit(x_train, y_train)
y_pred_elastic_net = elastic_net.predict(x_test)
print("Elastic Net R2 score:", r2_score(y_test, y_pred_elastic_net))
```

```python
class ElasticNetRegression:
    def __init__(self, alpha=0.001, l1_ratio=0.5, learning_rate=0.01, epochs=1000):
        self.alpha = alpha
        self.l1_ratio = l1_ratio
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.alpha_r = self.l1_ratio * self.alpha
        self.alpha_l = self.alpha - self.alpha_r

    def mode(self, x_train, y_train):
        n, f = x_train.shape
        self.coef_ = np.zeros(f)
        self.intercept_ = 0

        
        for _ in range(self.epochs):
            y_pred = self.predict(x_train)
            error = y_pred - y_train

            dw = ((2/n) * x_train.T.dot(error) + self.alpha_l * np.sign(self.coef_) + 2 * self.alpha_r * self.coef_)
            db = (2/n) * np.sum(error)

            self.coef_ -= self.learning_rate * dw
            self.intercept_ -= self.learning_rate * db

    def predict(self, x_test):
        return x_test.dot(self.coef_) + self.intercept_

```

```python
elastic_net_custom = ElasticNetRegression(alpha=0.001, l1_ratio=1, learning_rate=0.9, epochs=10000)
elastic_net_custom.mode(x_train, y_train)
y_pred_elastic_net_custom = elastic_net_custom.predict(x_test)
print("Custom Elastic Net R2 score:", r2_score(y_test, y_pred_elastic_net_custom))
```
