```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Simple input and output dataset

```python
# It ensures that the random numbers generated are the same every time the code is run. 42 isn't any special number.
reg = np.random.default_rng(seed=42)
x = reg.random((50, 1)) * 100
y = 3.5 * x + reg.random((50, 1)) * 20
```

Set up the model

```python
model = LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color="blue", label="Data Points")
plt.plot(x, y_pred, color="red", linewidth=2, label="Predicted line")
plt.title("Linear Regression on Random Dataset")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

Find slope and intercept

```python
print("Slope (coefficent): ", model.coef_[0][0])
print("Intercept: ", model.intercept_[0])
```
