```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
x = np.random.rand(50, 1) * 100
y = 3.5 * x + np.random.rand(50, 1) * 30
```

```python
class GDR:
    def __init__(self, lr, epochs):
        self.b = 0
        self.w = 0
        self.epochs = epochs
        self.lr = lr
        self.losses = []

    def model(self, x, y):
        for epoch in range(self.epochs):
            pred = self.w * x.ravel() + self.b
            err = np.mean(np.square(y.ravel() - pred))
            self.losses.append(err)

            # Gradients (scaled by number of samples)
            loss_slope_b = -2 * np.mean(y.ravel() - pred)
            loss_slope_w = -2 * np.mean((y.ravel() - pred) * x.ravel())

            # Update parameters
            self.b -= self.lr * loss_slope_b
            self.w -= self.lr * loss_slope_w

        # Plot loss curve
        plt.figure()
        plt.plot(self.losses, color="green")
        plt.title("Linear Regression - Loss (MSE)")
        plt.xlabel("Epochs")
        plt.ylabel("Loss (MSE)")
        plt.grid(True)

        # Plot fitted line
        plt.figure()
        plt.scatter(x.ravel(), y.ravel(), color="green")
        idx = np.argsort(x.ravel())
        plt.plot(x[idx], self.w * x[idx] + self.b, color="red", linewidth=2)
        plt.title("Linear Regression - Fitted Line")
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.grid(True)
        plt.show()
```

```python
gdr = GDR(0.00001, 75)
gdr.model(x, y)
```
